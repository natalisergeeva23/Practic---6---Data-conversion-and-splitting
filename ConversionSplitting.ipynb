Цель: Преобразовать признаки и разбить датасет на обучающую и тестовую выборку.

Для начала как всегда добавляем библиотеки и считываем данные с файла.

```python
import pandas as pd
import numpy as np
```

```python
#считывание данных с файла
df = pd.read_csv('vgsales_19.csv')
df
```

Продемонстрировать работу OrdinalEncoder и использовать его на датасете:
Подключаем специально для этого библиотеку с расширением OrdinalEncoder. Далее записывем в первую переменную координату. Указываем двумерный массив с указанием столбцов данных. Функция fit вычисляет формулировку для преобразования столбцов на основе стандартного масштабирования.

```python
from sklearn.preprocessing import OrdinalEncoder

Oenc = OrdinalEncoder()
Oenc.fit(df[['Platform','Publisher']])
df
```

Вызовем функцию "categories" для определения значений в записанной переменной. Например, в первой строке столбца Platform  стоит слово - DS, по списку оно под 3 номером. Произведем кодирование.

```python
Oenc.categories_
```

Обратимся к столбцам датасета и через функцию fit_transform, заменим словосочетания на цифры. Исходя из этого результата можно понять, что все правильно и DS находится под 3 номером, но тк счет идет с 0, переменная записалась под 2. 

```python
Oenc = OrdinalEncoder()
df[['Platform','Publisher']] = Oenc.fit_transform(df[['Platform','Publisher']])
df
```

Обязательно преобразовать 1 категориальный признак с помощью LabelEncoder.
Подключаем библиотку LabelEncoder. Данное кодирование одинаково с предыдущем, но различие все же есть. Различие в том, что данное кодирование преобразует целевую переменную. 

```python
from sklearn.preprocessing import LabelEncoder
Lenc = LabelEncoder()
Lenc.fit(df['Genre'])
df['Genre'] = Lenc.fit_transform(df['Genre'])
df
```

Видно как закодировались переменные. 

```python
Lenc.classes_
```

Провести нормализацию с помощью Normalize на числовых признаках.
Подключаем библиотеку Normalizer.
Принцип нормализации работает так - он вычитает минимальное значение из каждой записи, затем делит результат на разница между максимальным значением и минимальным значением. 
Передадим в переменную столбцы, которые будут нормализированы. 

```python
from sklearn.preprocessing import Normalizer

Norm = Normalizer()

df[['NA_Sales', 'EU_Sales']] = Norm.fit_transform(df[['NA_Sales', 'EU_Sales']])
df
```

Разделить датасет на обучающую и тестовую выборку.
Подключаем библиотеку train_test_split.
Y - столбец с обучающими данными, Х - столбец с тестовыми данными. Далле с помощью функции train_test_split разбиваем на выборки. 
random_state - параметр для управления перетасовкой,
test_size - значение с плавающей точкой. 
Посмотрим что находится в каждой переменной. 

```python
from sklearn.model_selection import train_test_split

Y = df['Year']
X = df.drop(columns=['Global_Sales'])

x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size=0.2, random_state=42)
```
